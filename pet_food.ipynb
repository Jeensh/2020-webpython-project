# 반려동물에게 위험한 음식을 알려주는 프로그램

### 음식 클래스

```python
class Food:
    name = ""
    foodType = ""
    ingredient = []
    peril = ""
    symptom = []
    note = ""
    
    def __init__(self, name = "", foodType = "", ingredient = [], peril = "", symptom = [], note= ""):
        self.name = name
        self.foodType = foodType
        self.ingredient = ingredient
        self.peril = peril
        self.symptom = symptom
        self.note = note
    
    def __str__(self):       
        rep = '-'*130 + "\n음식명 : {0}\n분류 : {1}\n위험도 : {2}\n위험재료 : {3}\n증상 : {4}\n비고 : {5}\n".format(
                self.name, self.foodType, self.peril, self.ingredient, self.symptom, self.note) + '-'*130
        return rep
        
    def __eq__(self, other):
        return self.name == other.name
    
```

### 음식 목록 클래스

```python
class foodList:
    foodList = []
    
    def getFoods_from_file(self, fileName, pet):
        """ (string, string) -> 반환값 없음
        
        파일명과 애완동물의 이름을 파라미터로 입력받아 해당 동물에 관련된 엑셀파일을 읽고 내용을 
        멤버리스트인 foodList(list형)에 저장하는 함수이다.
        파일명은 엑셀파일의 이름에 해당하고 애완동물의 이름은 엑셀파일의 시트명에 해당한다.
        foodList는 원소로 Food 클래스의 객체를 가진다.
        """
        import openpyxl
        
        wb = openpyxl.load_workbook(fileName)
        ws = wb[pet]
        
        count = 0
        for r in ws.rows:
            if count == 0 :                           #첫번째 행 스킵
                count = count + 1
                continue
            name = r[0].value
            type = r[1].value
            ingredient_str = r[2].value
            if ingredient_str == None:
                ingredient = []
            else:
                ingredient = ingredient_str.split(',')
            peril = r[3].value
            symptom_str = r[4].value
            if symptom_str == None:
                symptom = []
            else:
                symptom = symptom_str.split(',')
            note = r[5].value
            
            food = Food(name, type, ingredient, peril, symptom, note)
            self.foodList.append(food)
            count = count + 1
        
        wb.close()  
    def clear(self):
        """ (파라미터 없음) -> (반환값 없음)
        멤버리스트인 foodList를 empty상태로 초기화 하는 함수이다.
        """
        self.foodList.clear()
```

### 메뉴 클래스

```python
class Menu:
    def getPetName(self):
        """(파라미터 없음) -> string
        
        애완동물의 이름(종류)을 입력받아 반환하는 함수이다.
        
        >>> getPetName()
        애완동물의 종류를 입력하세요(ex.강아지)(0.종료) : 강아지
        '강아지'
        """
        pet = input("애완동물의 종류를 입력하세요(ex.강아지)(0.종료) : ")
        return pet
    
    def getFoodName(self):
        """(파라미터 없음) -> string
        
        음식의 이름을 입력받아 반환하는 함수이다.
        
        >>> getFoodName()
        음식이름을 입력하세요(0.뒤로가기) : 초콜렛
        '초콜렛'
        """
        name = input("음식이름을 입력하세요(0.뒤로가기) : ")
        return name
    
    def getKategrie(self):
        """(파라미터 없음) -> string
        
        분류의 이름을 입력받아 반환하는 함수이다.
        
        >>> getKategrie()
        분류 이름을 입력하세요(0.뒤로가기) : 채소
        '채소'
        """
        kat = input("분류 이름을 입력하세요(0.뒤로가기) : ")
        return kat
    
    def checkPetName(self, pet):
        """ (string) -> bool
        
        애완동물의 이름이 pets.txt 파일에 존재하는지 확인하여 
        존재하면 True 존재하지 않으면 False를 반환하는 함수이다.
        ※(pets.txt) 파일에는 엑셀에 있는 시트명(동물종류)가 작성되어 있다.
        
        예시: pets.txt에 강아지만 존재한다고 가정
        >>> checkPetName('강아지')
        True
        >>> checkPetName('고양이')
        해당 애완동물에 대한 정보는 없습니다.
        """
        L = []
        with open('pets.txt','r') as file:
            for p in file.readlines():
                L.append(p.strip())
        if(pet not in L):
            print("해당 애완동물에 대한 정보는 없습니다.")
            return False
        else:
            return True
        
    def petMode(self, fileName, pet):
        """ (string, stirng) -> 반환값 없음
        파일명과 애완동물의 이름(종류)를 파라미터로 받아 해당 동물의 엑셀 파일 내용을 읽고 저장한 후
        해당 동물에 대한 모드를 진행하는 함수이다.
        """
        foods = foodList()
        
        foods.getFoods_from_file(fileName, pet)
        
        while(True):
            print('1.음식 검색\n2.카테고리\n0.뒤로가기')
            key = int(input("키워드를 입력하세요 : "))
                      
            if key == 1:
                self.searchFood(foods)
            elif key == 2:
                self.kategorie(foods)
            elif key == 0:
                foods.clear()
                break
            else:
                print('잘못된 키워드를 입력하셨습니다')
                print('-'*50)
                continue     
                
    def searchFood(self, foodList):
        """ (foodList) -> 반환값 없음
        
        foodList클래스의 객체를 파라미터로 받아 음식 검색기능을 진행하는 함수이다.
        함수 시작시 음식의 이름을 입력으로 받고 foodList의 멤버 리스트에 입력으로 받은 음식의 이름과
        같거나 포함된 이름을 가진 음식의 목록을 출력한다.
        이후 입력으로 출력된 음식목록에 있는 음식의 이름을 입력하면 해당 음식에 대한 정보가 출력된다.
        """
        while(True):
            name = self.getFoodName()
            selectedFood = ""
            foundList = []
            index = 1
            if name == '0':
                break
            
            for f in foodList.foodList:
                if name in f.name:
                    print(str(index) + "." + f.name)
                    foundList.append(f)
                    index = index+1
            
            if index == 1:
                print("검색된 내용이 없습니다.")
            else:
                while(True):
                    found = False
                    selectedFood = self.getFoodName()
                    if selectedFood == "0":
                        break
                
                    for f in foundList:
                        if selectedFood == f.name:
                            print(f)
                            found = True
                            break
                    if not found :
                        print("목록에 입력한 음식이 없습니다.")
            
            if selectedFood == "0":
                break
            
    def kategorie(self,foodList):
        """ (foodList) -> 반환값 없음
        foodList클래스의 객체를 파라미터로 받아 카테고리 검색기능을 진행하는 함수이다.
        함수시작 시 foodList에 저장된 Food 클래스 객체의 '분류'를 출력하고 해당 분류의 이름을 
        입력으로 받는다.
        이후 입력으로 받은 분류에 해당하는 모든 Food 클래스 객체(음식)의 이름을 출력하고 
        음식의 이름을 입력받는다.
        이후 입력받은 음식의 정보를 출력한다.
        """
        while(True):
            typeSet = set()
            typeList = []
            index = 1
            for f in foodList.foodList:
                typeSet.add(f.foodType)
            
            for t in typeSet:
                if t == None:
                    continue
                typeList.append(t)
            typeList = sorted(typeList)
            
            for t in typeList:
                print(str(index) + '.' + t)
                index = index + 1
            
            kategorie = self.getKategrie()
            
            if kategorie == '0':
                break
            
            selectedType = []
            for f in foodList.foodList:
                if kategorie == f.foodType:
                    selectedType.append(f)
            
            if selectedType == []:
                print("해당 카테고리가 목록에 없습니다.")
            else:
                index = 1
                for f in selectedType:
                    print(str(index) + '.' + f.name)
                    index = index + 1
                
                while(True):
                    found = False
                    selectedFood = self.getFoodName()
                    if selectedFood == "0":
                         break
                
                    for f in selectedType:
                        if selectedFood == f.name:
                            print(f)
                            found = True
                            break
                    if not found :
                        print("목록에 입력한 음식이 없습니다.")
```

### 메인 프로그램

```python
import sys
menu = Menu()

while(True):
    pet = menu.getPetName()
    if(pet == '0'):
        break
    elif(not menu.checkPetName(pet)):
        continue
    menu.petMode('data.xlsx', pet)
```
